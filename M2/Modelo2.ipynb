```python
cs={'gender':{'m':1,'f':0}}
datos.replace(cs,inplace=True)
```

ENTRENE


```python
datos_entrenamiento=datos.sample(frac=0.8, random_state=0)
datos_test=datos.drop(datos_entrenamiento.index)
```

```python
etiquetas_entrenamiento = datos_entrenamiento.pop('time')
etiquetas_test=datos_test.pop('time')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo=LinearRegression()

modelo.fit (datos_entrenamiento, etiquetas_entrenamiento)
```

```python
predicciones=modelo.predict(datos_test)
```

```python
import numpy as np

from sklearn.metrics import mean_squared_error

error=np.sqrt(mean_squared_error(etiquetas_test, predicciones))
print("Error porcentual")
print(error*100)
```
